One Hot Encoding of the categorical Data

```python
def one_hot_encoding(df,columnName,keys):
  dfCol = list(df[columnName])
  df = df.drop(columns=[columnName])
  othersFlag = False
  if keys[-1] == 'Others':
    keys = keys[:-1]
    othersFlag = True
  print(othersFlag)
  for uniqueKey in keys[:-1]:
    newKeyCol = np.array([])
    for entry in dfCol:
      if entry == uniqueKey:
        newKeyCol = np.append(newKeyCol,1)
      else:
        newKeyCol = np.append(newKeyCol,0)
    df[uniqueKey] = newKeyCol
  newKeyCol = np.array([])
  if othersFlag == True:
    for entry in dfCol:
      if entry not in keys:
        newKeyCol = np.append(newKeyCol,1)
      else:
        newKeyCol = np.append(newKeyCol,0)
    df[columnName+'Others'] = newKeyCol
  return df
```

```python
df2 = one_hot_encoding(df,'Summary',['Mostly Cloudy','Partly Cloudy','Others'])
print(df2.head())
```

```python
X = dff.values[:, 1:]
y = dff.values[:, 0]
m = len(y)
```

```python
df.shape
```

```python
def feature_normalize(X):
  mu = np.mean(X, axis = 0)
  sigma = np.std(X, axis= 0, ddof = 1)
  X_norm = (X - mu)/sigma
  return X_norm, mu, sigma
```

```python
X, mu, sigma = feature_normalize(X)
```

```python
print('mu= ', mu)
print('sigma= ', sigma)
```

```python
X = np.hstack((np.ones((m,1)), X))
```

```python
def compute_cost(X, y, theta):
  predictions = X.dot(theta)
  errors = np.subtract(predictions, y)
  sqrErrors = np.square(errors) 
  J = 1/(2 * m) * errors.T.dot(errors)
  return J
```

```python
def gradient_descent(X, y, theta, alpha, iterations):
  cost_history = np.zeros(iterations)
  for i in range(iterations):
    predictions = X.dot(theta)
    #print('predictions= ', predictions[:5])
    errors = np.subtract(predictions, y)
    #print('errors= ', errors[:5])
    sum_delta = (alpha / m) * X.transpose().dot(errors);
    #print('sum_delta= ', sum_delta[:5])
    theta = theta - sum_delta;
    cost_history[i] = compute_cost(X, y, theta)
  return theta, cost_history
```

```python
theta = np.zeros(11)
iterations = 400;
alpha = 0.15;
```

```python
theta, cost_history = gradient_descent(X, y, theta, alpha, iterations)
print('Final value of theta =', theta)
print('First 5 values from cost_history =', cost_history[:5])
print('Last 5 values from cost_history =', cost_history[-5 :])
```

```python
plt.plot(range(1, iterations +1), cost_history, color ='blue')
plt.rcParams["figure.figsize"] = (10,6)
plt.grid()
plt.xlabel("Number of iterations")
plt.ylabel("cost (J)")
plt.title("Convergence of gradient descent")
```

```python
iterations = 400;
theta = np.zeros(11)

alpha = 0.005;
theta_1, cost_history_1 = gradient_descent(X, y, theta, alpha, iterations)

alpha = 0.01;
theta_2, cost_history_2 = gradient_descent(X, y, theta, alpha, iterations)

alpha = 0.02;
theta_3, cost_history_3 = gradient_descent(X, y, theta, alpha, iterations)

alpha = 0.03;
theta_4, cost_history_4 = gradient_descent(X, y, theta, alpha, iterations)

alpha = 0.15;
theta_5, cost_history_5 = gradient_descent(X, y, theta, alpha, iterations)

plt.plot(range(1, iterations +1), cost_history_1, color ='purple', label = 'alpha = 0.005')
plt.plot(range(1, iterations +1), cost_history_2, color ='red', label = 'alpha = 0.01')
plt.plot(range(1, iterations +1), cost_history_3, color ='green', label = 'alpha = 0.02')
plt.plot(range(1, iterations +1), cost_history_4, color ='yellow', label = 'alpha = 0.03')
plt.plot(range(1, iterations +1), cost_history_5, color ='blue', label = 'alpha = 0.15')

plt.rcParams["figure.figsize"] = (10,6)
plt.grid()
plt.xlabel("Number of iterations")
plt.ylabel("cost (J)")
plt.title("Effect of Learning Rate On Convergence of Gradient Descent")
plt.legend()
```

POLYNOMIAL REGRESSION

```python
Xt = X.T
Xp = Xt.copy()
for i in range(len(Xt)):
  Xp = np.vstack([Xp,Xt[i]**2])
  Xp = np.vstack([Xp,Xt[i]**3])
for i in range(len(Xt)):
  for j in range(i+1,len(Xt)):
    Xp = np.vstack([Xp,Xt[i]*Xt[j]])
for i in range(len(Xt)):
  for j in range(i+1,len(Xt)):
    for k in range(j+1,len(Xt)):
      Xp = np.vstack([Xp,Xt[i]*Xt[j]*Xt[k]])
```
